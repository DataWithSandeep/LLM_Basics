#### When to use create_documents(), Split_text(), split_documents()
* When working with text loaders and text splitters, the create_documents() method helps convert split text into LangChain's Document objects with optional metadata. It Converts a list of text chunks into Document objects.
Allows adding metadata (e.g., filename, source, timestamp). Helps in storing & retrieving documents efficiently.
1. **split_text**  
   - Splits a single text string into chunks.  
   - Use when working with plain text (e.g., articles, books).  
   - Example: `splitter.split_text(text)`

2. **split_documents**  
   - Splits a list of `Document` objects while preserving metadata.  
   - Use when processing structured documents (e.g., PDFs, research papers).  
   - Example: `splitter.split_documents(docs)`

3. **create_document**  
   - Creates a `Document` object from text with metadata.  
   - Useful for formatting raw text before processing.  
   - Example: `splitter.create_document(text, metadata={"source": "book"})`
**Key Difference:** `split_text` works with raw text, while `split_documents` ensures metadata is preserved.  



#### üìå Difference Between load() and load_and_split() in LangChain File Loaders.

* load() ‚Üí Loads the file as a single Document object.
* load_and_split() ‚Üí Loads the file and automatically splits it into multiple Document objects using a text splitter.
* If you use load_and_split() without specifying a text splitter, LangChain automatically applies the RecursiveCharacterTextSplitter (RCT) with default parameters
* example:
* loader = PyPDFLoader("sample.pdf")
* splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)  # ‚úÖ Use a text splitter (adjust chunk size as needed)
* documents = loader.load_and_split(splitter) # ‚úÖ Load and split the document
* documents = loader.load_and_split() -- In this case it use by default RCT with default parameter.`

### Vector Stores (aka. Vector Databases)
* Store embeddings in a very fast searchable database.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_community.document_loaders import PyPDFLoader
# from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter
from langchain_chroma import Chroma

# Load the document, split it into chunks, embed each chunk and load it into the vector store.
loaded_document = PyPDFLoader(r'C:\Users\Sandeep.C\Downloads\TheShortStory_10001379.pdf').load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

chunks_of_text = text_splitter.split_documents(loaded_document)

vector_db = Chroma.from_documents(chunks_of_text, embedding)
```

```python
question = "There is no finer model for the study of setting than this story affords "
# response = vector_db.similarity_search_with_score(question,k=3)
response = vector_db.similarity_search(question,k=3)
print(len(response))
```

```python
print(response)
```

The `.similarity_search` method in the previous code is used to find the most relevant text chunks from a pre-processed document that are similar to a given query. Here's how it works step-by-step

1. **Using `.similarity_search`:**
   - When you invoke `vector_db.similarity_search(question)`, the method converts the query (`questions`) into an embedding using the same method that was used for the chunks.
   - It then searches the vector database (`vector_db`) to find the chunks whose embeddings are most similar to the embedding of the query. This similarity is typically measured using metrics such as `cosine similarity`. Rembeber that vector_db which is knowledge base in this case has 2 type of chunk search `.similarity_search` and `as_retriver()`
   - The search results are sorted by relevance, with the most relevant chunks (those that are semantically closest to the query) returned first.

2. **Output:**
   - The result of the `.similarity_search` is stored in `response`, which contains the relevant chunks and their similarity scores.
   - The script prints the content of the most relevant chunk (the first result), which should ideally contain information about what the president said regarding the John Lewis Voting Rights Act.

This method is particularly useful in applications like question answering or document retrieval where you need to quickly find the most relevant parts of a large text based on a query. an another thing is *similarity_search_with_score* it will return matching score along with metadata.

`as_retriever()` Converts a vector store (or another retriever) into a retriever object that can be used in LangChain pipelines.Used when integrating with chains and agents.When you need a retriever-compatible object for chaining with LangChain tools like ConversationalRetrievalChain. By default it is also uses similarity search in backend but you can custimize it, that the key difference.
Key Difference
* similarity_search() is a direct method to retrieve documents based on similarity.
* as_retriever() wraps the vector store into a retriever interface for use in LangChain pipelines.

BY default similarity_search return single relavent chunk of docs we can specify no of documnets we want as through K parameter. while as_retriever() by default return top 4 similar docs. here also we can specify no of relavent docs we want.
Also we can customize the retriever behavior when calling as_retriever(), such as adjusting the search method and retrieval strategy.retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={"k": 5})
search_type="mmr": Uses Maximal Marginal Relevance (MMR) instead of similarity search.
search_kwargs={"k": 5}: Retrieves the top 5 most relevant documents.

```python
retriever = vector_db.as_retriever()
```

```python
response=retriever.invoke('There is no finer model for the study of setting than this story affords')
```

* Difference Between retriever.get_relevant_documents() and retriever.invoke() in LangChain.
* Both retriever.get_relevant_documents() and retriever.invoke() are used to retrieve documents, but they have different use cases and behaviors.
* Use get_relevant_documents() if you need a simple, direct document retrieval. It takes a string query and searches for similar documents. does't support LCEL.
* Use invoke() if you're working with LCEL or need to integrate the retriever into a chain or agent-based workflow.Can accept different input types (dict, tuples, structured prompts).

```python
response=retriever.get_relevant_documents('There is no finer model for the study of setting than this story affords')
```
